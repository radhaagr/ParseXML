```python
d = './train/'

alldir = [os.path.join(d, o) for o in os.listdir(d) 
                    if os.path.isdir(os.path.join(d,o))]
```

```python
#print(alldir)
```

```python
allfiles = []
for mypath in alldir:
    curfiles = [os.path.join(mypath, f) for f in os.listdir(mypath) if os.path.isfile(os.path.join(mypath, f))]
    allfiles.extend(curfiles)
```

```python
#print(allfiles)
```

```python
import xml.etree.ElementTree as ET
from lxml import etree
import itertools
#!pip install xmltodict
```

```python
from bert_embedding import BertEmbedding
import pandas as pd
import numpy as np
import mxnet as mx
import pickle
```

```python
ctx = mx.gpu(0)
bert = BertEmbedding(ctx=ctx)
```

```python
def label_polarity(rating):
    if rating == 3.0:
        return 'neutral'
    elif rating > 3.0:
        return 'positive'
    else:
        return 'negative'
```

```python
def label_polarity_num(rating):
    if rating == 3.0:
        return 1
    elif rating > 3.0:
        return 2
    else:
        return 0
```

```python
utf8_parser = etree.XMLParser(encoding='utf-8', recover=True)
x_features = []
y = []
with open("./train/Reviews_Task2.txt",'w',encoding = 'utf-8') as ftext, open("./train/polarity.txt",'w',encoding = 'utf-8') as flabel:
    for eachfile in allfiles:
        with open(eachfile) as f:
            print(eachfile)
            it = itertools.chain('<root>',f,'</root>')
            root = ET.fromstringlist(it,parser=utf8_parser)
            for child in root:
                if child.tag in 'STARS':
                    flabel.write(label_polarity(float(child.text))+'\n')
                    y.append(label_polarity_num(float(child.text)))
                if child.tag in 'REVIEW':
                    #print(len(child.text))
                    ftext.write(str(child.text).strip().strip('\n')+'\n\n')
                    result = bert.embedding([child.text])
                    #print("length",len(result))
                    for i in np.arange(len(result)):
                        x_features.append(result[i][0])
            
x_features = np.array(x_features)
print("features shape",x_features.shape)
np.save('bert_sample_features',x_features)
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test = train_test_split(x_features,y,test_size=0.2,random_state=47)
```

```python
import lightgbm as lgb
```

```python
lgb_model = lgb.LGBMClassifier()
lgb_model.fit(x_train,y_train)
y_pred = lgb_model.predict(x_test)
```

```python
# to save model
filename = 'Task_2_model'
pickle.dump(lgb_model, open(filename, 'wb'))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
print(cross_val_score(lgb_model,x_features,y,cv=5))
```
